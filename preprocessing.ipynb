```python
# Load and clean data
df = pd.read_excel("C:\\Users\\yogya\\Desktop\\IPL Data\\IPL 2025  league stage.xlsx")

def clean_team_name(name):
    if pd.isna(name): return name
    name = str(name).strip()
    return 'PBKS' if name == 'PKBS' else name

df['home team'] = df['home team'].apply(clean_team_name)
df['visiting team'] = df['visiting team'].apply(clean_team_name)
df['results'] = df['results'].apply(clean_team_name)

# Refined Importance Function
def get_refined_importance(current_pts, matches_played):
    total_m = 14
    m_rem = total_m - matches_played
    if m_rem <= 0: return 0.0

    # Progress Weight: 0.5 at start of season -> 1.0 at final match
    progress = matches_played / total_m
    weight = 0.5 + (0.5 * progress)

    # Goal 1: Top 4 (16 pts)
    needed_4 = 16 - current_pts
    imp_4 = (needed_4 / (2 * m_rem)) if 0 < needed_4 <= (2 * m_rem) else 0
    
    # Goal 2: Top 2 (20 pts)
    needed_2 = 20 - current_pts
    imp_2 = (needed_2 / (2 * m_rem)) if 0 < needed_2 <= (2 * m_rem) else 0
    
    # Selection logic: Primary is Top 4. If qualified for Top 4, pursue Top 2.
    # We weight the pursuit of Top 2 slightly lower than the 'survival' of Top 4.
    raw_imp = max(imp_4, imp_2 * 0.7)
    final_imp = raw_imp * weight
    
    # Floor: Dominant teams already at 20+ points (Maintenance/Momentum)
    if current_pts >= 20:
        return round(0.15 + (m_rem * 0.02), 3)
    
    # Floor: Mathematically eliminated teams (Pride/Experimentation)
    if needed_4 > 2 * m_rem:
        return round(0.10 + (m_rem * 0.01), 3)

    return round(min(final_imp, 1.0), 3)

# Tracking state through the season
teams = pd.concat([df['home team'], df['visiting team']]).unique()
points = {team: 0 for team in teams if pd.notna(team)}
matches_played = {team: 0 for team in points}

home_imps, visitor_imps = [], []

for idx, row in df.iterrows():
    home, visitor, result = row['home team'], row['visiting team'], row['results']
    
    # Calculate pre-match importance
    home_imps.append(get_refined_importance(points[home], matches_played[home]))
    visitor_imps.append(get_refined_importance(points[visitor], matches_played[visitor]))
    
    # Update points for next match calculations
    matches_played[home] += 1
    matches_played[visitor] += 1
    if result == home: points[home] += 2
    elif result == visitor: points[visitor] += 2
    elif result in ['no result', 'tie']:
        points[home] += 1
        points[visitor] += 1

# Add to dataframe
df['home_importance'] = home_imps
df['visitor_importance'] = visitor_imps
df.to_csv('IPL_2025_refined_importance.csv', index=False)
```
