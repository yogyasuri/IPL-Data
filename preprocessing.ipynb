```python
def overs_to_balls(x):
    overs = int(x)
    balls = round((x - overs) * 10)  # because .2 means 2 balls
    return overs * 6 + balls
```

```python
bowling['balls'] = bowling['Overs'].apply(overs_to_balls)
```
